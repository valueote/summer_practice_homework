```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_excel('loandata.xls')
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
X_Data = data.drop(['违约'],axis=1)
X_Data.head()
```

```python
y_data = np.ravel(data[['违约']])
y_data[0:5]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel

logistic = LogisticRegression(max_iter=1000)
selector = SelectFromModel(estimator=logistic)
selector.fit(X_Data, y_data)
X_selected = selector.transform(X_Data)
logistic.fit(X_selected, y_data)
score = logistic.score(X_selected, y_data)
print(f"Model accuracy: {score:.2f}")
```

```python
X2_data = data.drop(['年龄','教育','收入','其他负债','违约'],axis=1)
X2_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X2_data,y_data,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train , y_train)
```

```python
lr.coef_
```

```python
 lr.intercept_
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
