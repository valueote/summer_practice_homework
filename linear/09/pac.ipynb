```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# Creating example data
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T

plt.scatter(X[:, 0], X[:, 1])
plt.axis("equal")
plt.show()
```

```python
# Importing PCA from Scikit-Learn
from sklearn.decomposition import PCA

# Creating PCA model and fitting it to data X
pca = PCA(n_components=2)
pca.fit(X)

# Viewing the principal components
print(pca.components_)

# Checking the shape of components_
print(pca.components_.shape)

# Checking the type of pca.components_
print(type(pca.components_))

# Explained variance of each principal component
print(pca.explained_variance_)
```

```python
# Reducing components to 1
pca = PCA(n_components=1)
pca.fit(X)

# Transforming data X to X_pca
X_pca = pca.transform(X)

# Printing shapes of original and transformed data
print("original shape:", X.shape)
print("transformed shape:", X_pca.shape)

# Slicing to print first ten items of X and X_pca
print(X[:10])
print(X_pca[:10])
```

```python
# Inverse transforming reduced data to original dimensions
X_new = pca.inverse_transform(X_pca)

# Plotting original data and reconstructed data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal')
plt.show()
```

```python
# Importing dataset and checking its shape
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)

# Training PCA on digits.data and visualizing cumulative explained variance
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("number of components")
plt.ylabel("cumulative explained variance")
plt.show()
```
